```python
# ライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
import statistics
import os
import itertools
import math
import seaborn as sns
import matplotlib.cm as cm # グラデーション用

%matplotlib inline
```

### Loading Data

```python
filename_list = ["5", "10", "15"]

df_list = [pd.read_csv(f"/Users/hidaka/Documents/Wolfram Mathematica/20240609_outputs_c/result/M2={filename_list[i]}.csv") for i in range(len(filename_list))]
df_list[0]
```

### Change in ts when mub2/mub1 is changed

```python
# 横軸：c
x_list = [df_list[i]["c"] for i in range(len(df_list))]

# 縦軸：ts
y_list = [df_list[i]["ts"] for i in range(len(df_list))]

# 色
color_list = []

# 可視化
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)

for i in range(len(df_list)):
    ax.plot(x_list[i], y_list[i], linewidth=4, label=f"$M_{2}$={filename_list[i]}", color=cm.Blues((i+0.7)/len(df_list)))

ax.set_xlabel("Advantage of late male, $c$", fontsize=18, labelpad=10)
ax.set_ylabel("Switching time bitween \n early male and late male, $t_{\mathrm{s}}$", fontsize=18, labelpad=20)
ax.tick_params(direction='in', which='both', bottom=True, top=True, left=True, right=True, width=1.5)  # 目盛りを内側にする
ax.spines['bottom'].set_linewidth(1.5)  # 下の枠の太さを変更
ax.spines['top'].set_linewidth(1.5)     # 上の枠の太さを変更
ax.spines['left'].set_linewidth(1.5)    # 左の枠の太さを変更
ax.spines['right'].set_linewidth(1.5)   # 右の枠の太さを変更
ax.tick_params(axis='both', which='major', labelsize=12)  # メモリの数値の大きさを変更
# ax.grid(True)
ax.legend(bbox_to_anchor=(1, 0.85), loc='upper right', borderaxespad=1, fontsize=16, frameon=False)

plt.savefig('result_c.png', dpi=300, bbox_inches='tight')  # dpiは解像度、bbox_inches='tight'は画像の余白を削除
```

### Changes in C1 and C2

```python
# 横軸：c
x_list = [df_list[i]["c"] for i in range(len(df_list))]

# 縦軸：C1,C2
y1_list = [df_list[i]["C1"] for i in range(len(df_list))]
y2_list = [df_list[i]["C2"] for i in range(len(df_list))]

# 可視化
# C1
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)
for i in range(len(df_list)):
    ax.plot(x_list[i], y1_list[i], linewidth=3, label=f"M2={filename_list[i]}")
ax.set_xlabel("c")
ax.set_ylabel("C1")
ax.grid(True)
ax.legend()

# C2
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)
for i in range(len(df_list)):
    ax.plot(x_list[i], y2_list[i], linewidth=3, label=f"M2={filename_list[i]}")
ax.set_xlabel("c")
ax.set_ylabel("C2")
ax.grid(True)
ax.legend()
```

### Lambda Changes

```python
# 横軸：c
x_list = [df_list[i]["c"] for i in range(len(df_list))]

# 縦軸：λ
y_list = [df_list[i]["Lambda"] for i in range(len(df_list))]

# 可視化
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)

for i in range(len(df_list)):
    ax.plot(x_list[i], y_list[i], linewidth=3, label=f"M2={filename_list[i]}")

ax.set_xlabel("c")
ax.set_ylabel("λ")
ax.grid(True)
ax.legend()
```
